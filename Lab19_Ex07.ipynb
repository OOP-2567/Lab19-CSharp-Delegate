## Lambda expression



1.สร้าง console application project

```cmd
dotnet new console --name Lab19_Ex07
```



2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
// Lambda expression
Square getSquare = x=> x*x ; // 2. define lambda expression

int num = 9;
int sqr = getSquare(num);
System.Console.WriteLine($"Square value of {num} is {sqr}");

delegate int Square(int num);  // 1. declare delegate
```



3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab19_Ex07
```



6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5



7.อธิบายสิ่งที่พบในการทดลอง


ประกาศ delegate Square ซึ่งรับพารามิเตอร์เป็น int และคืนค่าผลลัพธ์เป็น int

กำหนด lambda expression x => x * x ซึ่งทำหน้าที่คำนวณค่าสแควร์ของตัวเลขที่รับเข้ามา (x), ซึ่งสามารถแทนที่ anonymous method ได้

การใช้ lambda expression: เมื่อเรียก getSquare(num) โดยที่ num เป็น 9, จะคำนวณผลลัพธ์เป็น 81 และแสดงผลในคอนโซล
